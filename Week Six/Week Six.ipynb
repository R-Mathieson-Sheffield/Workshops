### Plotting
We are going to use python to plot our data, this will allow us to follow trends in data a lot easier. Python can not intrinsically plot to the screen only text to a console. In order to plot data we are going to need to load in a library. Exactly the same as when we used numpy.
The library we need is called matplotlib, it was originally based on the plotting library available in matlab but has become a very popular and powerful plotting library. Let us import it.

```python
#Importing matplotlib
import matplotlib.pyplot as plt
```

One thing to notice is that we import a sub-library, called pyplot from the main library called matplotlib. Let us use it to look at our data.

```python
#Example
plt.plot(X, ySin) # we use the plot command with two arguments, the X data and the Y data.
plt.plot(X, yCos) # If we want more than one figure we just add another plot.
# How easy is that.
# If you are using the command line or an IDE you will also need to add the command
#plt.show()
```

A quick recap:
We define a figure to which we can add stuff.
Then we add some axis to the figure, and then we plot in to that axis.  Notice that the axis were defined to be between 0.5 and 1.0, but the plot co,,and overwrites this preference, if we want to zoom in then we have to change the axis after the plot command

```python
fig = plt.figure() # Notice that we have to define the figure again, all the processing of a figure needs to happen in a single cell, because at the end of the cells execution matplotlib expects to present the plot
axes = fig.add_axes([0.5, 1, 0.5, 1])
axes.plot(X,ySin)
axes.axis([0.75, 2.4, 0.7, 1]) # here I can set the scale of the axis
axes.set_title("zoomed in")
axes.set_xlabel('Angle')
axes.set_ylabel('Value')
```

### Multiple plots and plot characters
Now we will look at putting multiple axis on a single figure and how we can customise the plots.

```python
fig = plt.figure()              # Again we build a figure
ax_1=fig.add_subplot(1, 2, 1)   # We add a subplot taking three arguments, the first is the number of plots in Y and the second is the number in the X direction, the final number is which of the axis the one we are defining is.

# We define a subplot which belongs to a group 1 high and 2 across, this is the first onw
ax_2=fig.add_subplot(1, 2, 2)   # This is the second one.

# Let us pot in to them. for the first plot
ax_1.plot(X, ySin)
ax_1.axis([0, 6.28, -1, 1])
ax_1.set_title("Sine")
ax_1.set_xlabel('Angle')
ax_1.set_ylabel('Value')

#For the second plot
ax_2.plot(X, yCos)
ax_2.axis([0, 6.28, -1, 1])
ax_2.set_title("Cosine")
axes.set_xlabel('Angle')
ax_2.set_ylabel('Value')
```

We can see that this produces a nice graph.  However, there are some problems.
The spacing is a little out, and it would be nice to have a little more colour and bolder text.  Let us correct that now.

```python
fig = plt.figure()              # Again we build a figure
ax_1=fig.add_subplot(1, 2, 1)   # We add a subplot taking three arguments, the first is the number of plots in Y and the second is the number in the X direction, the final number is which of the axis the one we are defining is.

# We define a subplot which belongs to a group 1 high and 2 across, this is the first onw
ax_2=fig.add_subplot(1, 2, 2)   # This is the second one.

# Let us pot in to them. for the first plot
ax_1.plot(X, ySin, 'r')  # Here I use the string 'r' to change the colour to red
ax_1.axis([0, 6.28, -1, 1])
ax_1.set_title("Sine", fontsize=32, fontweight="bold") # I can change the font size and makie it bold
ax_1.set_xlabel('Angle', fontsize=14)
ax_1.set_ylabel('Value', fontsize=14)

#For the second plot
ax_2.plot(X, yCos, 'g o') # Here I use the string 'g o', g is for green and o is for circles
ax_2.axis([0, 6.28, -1, 1])
ax_2.set_title("Cosine", fontsize=32, fontweight="bold")
ax_2.set_xlabel('Angle', fontsize=14)
ax_2.set_ylabel('Value', fontsize=14)

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=1.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4) # This command allow us to control the spacing between plot
```

### Plotting fitted data
Let us generate some data

```python
gradientData = 3
offsetData = 4
noise = 10
numberPoints = 100
xData = np.linspace(0,numberPoints-1,numberPoints)
yData = offsetData + xData * gradientData + np.random.normal(0,noise, numberPoints)
plt.plot(xData, yData)
```

### Lets fit that data

```python
from scipy import stats
gradient, intercept, rV, PV, stderr = stats.linregress(xData, yData)
print(gradient)
print(intercept)
```

### We can generate some new data from our fit parameters and plot it.

```python
## I make my x values, I will have 10 this time
xFit = np.linspace(0, 100, 10)
yFit = gradient * xFit + intercept
```

### Lets plot those two data sets

```python
plt.plot(xData, yData, 'o')
plt.plot(xFit, yFit, 'r')
```

## Other types of plots
We will now look to creating some other plot types, these are less commonly used in physics.  However, they may be of use to you

##### Histogram

```python
# Example of a histogram
testData = np.random.normal(0, 10, 1000) # I am using a numpy command to generate a guassian data set.
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
myBins = np.linspace(-20, 20, 41)
axes.hist(testData, bins=30, color='g') # Here you can have a set number of bins or supply them with bins=myBins. A bin is the region in which the data is sorted to make the histogram.

# From the documention
#If bins is an integer, it defines the number of equal-width bins in the range.
#
#If bins is a sequence, it defines the bin edges, including the left edge of the first bin and the right edge of the last bin; in this case, bins may be unequally spaced.
```

##### Pie chart

```python
# Example of pie chart
# The pie chart needs to know the size of each
labels = 'Bluey', 'Bingo', 'Chilli', 'Bandit'
sizes = [40, 30, 10, 20]

fig = plt.figure()
axes = fig.add_axes([0.5, 1, 0.5, 1])
axes.pie(sizes, labels=labels, shadow=True, startangle=90)
axes.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
```

```python
# Example of pie chart
# The pie chart needs to know the size of each
labels = 'Bluey', 'Bingo', 'Chilli', 'Bandit'
sizes = [40, 30, 10, 20]
exploded = [.1, 0, 0, 0] # This is an array which tells matplotlib to move the pie slice out by

fig = plt.figure()
axes = fig.add_axes([0.5, 1, 0.5, 1])
axes.pie(sizes, labels=labels, explode = exploded, shadow=True, startangle=90)
axes.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
```

```python
import matplotlib.gridspec as gridspec # We need another library
fig = plt.figure() # Start with a blank figure

gs0 = gridspec.GridSpec(1, 2, figure=fig) # Here we tell matplotlib we want two grid regions
# The two regions are gs0[0] and gs0[1]. It is just a list of places to put data
# this looks like this [gs0[0]] [gs0[1], exactly the same as subplot


gs00 = gs0[0].subgridspec(3, 3) # this gives me a 3 x 3 block in th region gs0[0]
# gs0[0] now is broken up in to 3 x3 [] [] []
#                                    [] [] []
#                                    [] [] []

ax1 = fig.add_subplot(gs00[:-1, :]) # Goes from the top to the last but one all the way across
ax1.text(0.3,0.5,'This is the region gs00[:-1, :]')
ax2 = fig.add_subplot(gs00[-1, :-1])
ax3 = fig.add_subplot(gs00[-1, -1])


gs01 = gs0[1].subgridspec(3, 3)
ax1 = fig.add_subplot(gs01[0, :])
ax2 = fig.add_subplot(gs01[1, :])
ax3 = fig.add_subplot(gs01[2, :])





# Still eed to correct the spaces
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=1.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
```

```python
# Now an example
fig = plt.figure(figsize=(8, 8)) # Start with a blank figure
gs0 = gridspec.GridSpec(1, 1, figure=fig)
gs00 = gs0[0].subgridspec(3, 3)

ax1 = fig.add_subplot(gs00[0:2, 0:2])
ax1.imshow(imagedata)


ax2 = fig.add_subplot(gs00[2:3, 0:2])
ax2.plot(imagedata[50,:])
ax3 = fig.add_subplot(gs00[0:2, 2:3])
ax3.plot(imagedata[:,50], np.linspace(0,1,100))
ax3.set_aspect(2)

plt.subplots_adjust(left=0.1,
                    bottom=0.3,
                    right=.8,
                    top=0.9,
                    wspace=0.1,
                    hspace=0.4)
```

```python
# using the scatter plot
# The scatter plot allows us to define the colour and shape of each point
# This is simply supplied as a list of the same size as the x data
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2 * np.pi, 100)
y = np.sin(x)
colours = []
for I in y:
    if I > 0:
        colours.append('red')
    else:
        colours.append('blue')
# Here colours is a list which contains either blue or red

sizes = 10*abs(y)

# Sizes is a list where the value of each data point is 10 * the absolute value of y

fig = plt.figure()
axes = fig.add_axes([0.5, 1, 0.5, 1])
axes.scatter(x,y, s=sizes, c=colours) # suplying s= the sizes and c = the colours
```
