```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
dataPoints = 500
offGradient = 2
onGradient = 5

X = np.linspace(0,499,500)
traceOne = np.zeros(dataPoints)
traceTwo = np.zeros(dataPoints)

traceOneOn = np.zeros(dataPoints)
traceOneOn[50:400] = 1
traceTwoOn = np.zeros(dataPoints)
traceTwoOn[0:200] = 1
traceTwoOn[400:500] = 1

noise = 5

for I in range(1,dataPoints):
    if traceOneOn[I] == 1:
        traceOne[I] = traceOne[I-1] + onGradient + np.random.normal(0,noise,1)
    else:
        traceOne[I] = traceOne[I-1] + offGradient + np.random.normal(0,noise,1)

    if traceTwoOn[I] == 1:
        traceTwo[I] = traceTwo[I-1] + onGradient + np.random.normal(0,noise,1)
    else:
        traceTwo[I] = traceTwo[I-1] + offGradient + np.random.normal(0,noise,1)

plt.plot(X, traceOne)
plt.plot(X, traceTwo)

myData = np.zeros([dataPoints,3])
myData[:,0] = X
myData[:,1] = traceOne
myData[:,2] = traceTwo

```

```python
np.savetxt('ProjectData/TraceData.csv', myData, delimiter=',',header="Time ms,TraceOne,TraceTwo")
```
