```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fileName = 'Data/results.csv'
data = pd.read_csv(fileName)
```

```python
teams = data['home_team'].unique()
```

# Can we calculate the most successful team, i.e. the team that has the highest winning percentage.
We do what we have just done before, but we do it for each team.
Below we will get the unique values for each team who played at home, We will use a for loop through that list.

```python
teamNames = []
percentageWins = []
for currentTeam in teams:
    currentData = data[(data['home_team']==currentTeam) | (data['away_team']==currentTeam)]
    HomeWins = currentData[(currentData['home_team'] == currentTeam) & (currentData['home_score'] > currentData['away_score'])]
    AwayWins = currentData[(currentData['away_team'] == currentTeam) & (currentData['away_score'] > currentData['home_score'])]
    teamNames.append(currentTeam)
    percentageWins.append( (HomeWins.shape[0] + AwayWins.shape[0]) / currentData.shape[0] )
```

```python
# I am going build a dataframe from these two lists
plotingData = pd.DataFrame({'Names': teamNames, '%': percentageWins})
plotingData.head
plotingData = plotingData.sort_values('%', ascending=False)
```

```python
plt.plot(plotingData['Names'][0:20], plotingData['%'][0:20])
plt.xticks(rotation = 90)
```
