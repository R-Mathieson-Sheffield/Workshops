# How to generate data
Here we are going to generate some data to fit.
We are going to use a very simple linear model

```python
def linear(Xdata,gradient, intercept):
    return intercept + gradient * Xdata
```

Great we have a simple function to generate a linear model.
Let's try it out

```python
import numpy as np
import matplotlib.pyplot as plt
numberOfPoints = 100
xdata = np.linspace(0,10,numberOfPoints)
ydata = linear(xdata, gradient=2, intercept=1)
plt.plot(xdata, ydata)
```

While that is brilliant, it would be very easy to fit. Lets make it harder fot the fitting function by adding some noise. We are going to add gaussian  noise, but you could use what ever is most suited to the physics.

```python
noise = np.random.normal(0,1,numberOfPoints)
plt.plot(noise)
```

We can add this to the data from the model, and plot it

```python
ydata = ydata + noise
plt.plot(xdata, ydata)
```

# Once you have fitted you data you will get back a matrix with the parameters in.

```python
popt = np.asarray([2,1])
```

# You can use this to plot the fit data as such

```python
fitdata = linear(xdata, *popt)
# Notice that I pass the array as *poopt
# The special syntax *args in function definitions in python is used to pass a variable number of arguments to a function.
plt.plot(xdata, ydata, 'o')
plt.plot(xdata, fitdata)
```
