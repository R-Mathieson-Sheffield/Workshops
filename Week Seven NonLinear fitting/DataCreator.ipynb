```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Exponential
# nt = no exp(-t/Tor)
x = np.linspace(0,10,100)
amplitude = 100
timeConst = 2
y = amplitude * np.exp(-(x/timeConst)) + np.random.normal(0,5,100)
data = np.asarray([x,y])
data = np.transpose(data)
plt.plot(data[:,0], data[:,1])
```

```python
np.savetxt("exp.csv", data, delimiter=",")
```

```python
# Power law
# y = offset + x^a
x = np.linspace(0,10,100)
offset = 1
power = 2.5
y = offset+ x**power + np.random.normal(0,10,100)
data = np.asarray([x,y])
data = np.transpose(data)
plt.plot(data[:,0], data[:,1])
```

```python
np.savetxt("power.csv", data, delimiter=",")
```

```python
# Gaussian
# nt = no exp(-t/Tor)
x = np.linspace(0,100,100)
centre = 30
width = 10
Ydata = amplitude * np.exp(-1*((centre-x)**2)/((width**2)))+ np.random.normal(0,5,100)
data = np.asarray([x,Ydata])
data = np.transpose(data)
plt.plot(data[:,0], data[:,1])
```

```python
np.savetxt("Gaussian.csv", data, delimiter=",")
```

```python
# sine wave law
# y = offset + x^a
x = np.linspace(0,3.14,100)
offset = 1
freq = 0.25
phaseOffest = 0.4
y = offset+ np.sin(x/freq + phaseOffest) + np.random.normal(0,.1,100)
data = np.asarray([x,y])
data = np.transpose(data)
plt.plot(data[:,0], data[:,1])
```

```python
np.savetxt("Sine.csv", data, delimiter=",")
```

# Lorenzian

```python
import numpy as np
def Lorentz(xdata, xo, gamma):
    ydata = (gamma/np.pi)/((xdata-xo)**2 + gamma**2)
    return ydata
```

```python
xdata = np.linspace(0,99,100)
ydata = Lorentz(xdata, 50, 8) + np.random.normal(0, .003, 100)
```

```python
data = np.asarray([xdata,ydata])
data = np.transpose(data)
plt.plot(data[:,0], data[:,1])
```

```python
np.savetxt("Lorentz.csv", data, delimiter=",")
```
