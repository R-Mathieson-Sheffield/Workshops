```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
gradient = 4
intercept = 1
y = gradient * x + intercept + np.random.normal(0, 5, 100)
```

```python
plt.plot(x,y)
```

```python
mydata = np.transpose(np.asarray([x,y]))
```

```python
np.savetxt("data1.csv", mydata, delimiter = ",")
```

```python
newdata = np.loadtxt("data1.csv", delimiter = ",")
```

```python
plt.plot(newdata[:,0], newdata[:,1])
```
