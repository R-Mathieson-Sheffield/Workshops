```python
import math
import numpy as np

def mycos(angle):
    mycosv = 1 - (angle ** 2) / math.factorial(2) + (angle ** 4) / math.factorial(4) - (angle ** 6) / math.factorial(6)
    return mycosv
```

```python
print(math.cos(3.14/4))
print(mycos(3.14/4))
```

```python
def numberMyCos(angle, number):
    holdall = 1
    sign = -1
    number =np.arange(2,number*2,2)
    for loopnumber in number:
        holdall = holdall + (sign * (angle**loopnumber)/math.factorial(loopnumber))
        sign = sign * -1
    return holdall
```

```python
print(math.cos(3.14/4))
print(mycos(3.14/4))
print(numberMyCos(3.14/4,40))
```

```python
import matplotlib.pyplot as plt
builting = math.cos(3.14/4)
data = []
for I in range(10):
    myanswer = numberMyCos(3.14/4,I)
    data.append(builting-myanswer)
plt.plot(data)
```

```python
import matplotlib.pyplot as plt
builting = math.cos(3.14/4)
numberOfLoops = 10
data = np.zeros(numberOfLoops)
for I in range(numberOfLoops):
    myanswer = numberMyCos(3.14/4,I)
    data[I] = builting-myanswer
plt.plot(data)
```
